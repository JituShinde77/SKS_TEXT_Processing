## Exercise 1
* Write a function that takes OfferDetails as input, and returns FaceValue as output
* Run the function against all rows in the attached data and determine function accuracy (Higher the better)


Hint: For Exercise 1, regex function should do (ensure that all edge cases are taken care of)

```python
c = ['','s', 's50','$$', '', '50$', '50', '$2.0','$2 50','$2 50 ','$2 50 50)))', '$2 50 50))) $$$$05,.m 50 150']
for i in c:
    dd = re.findall(r"\$\d+(?:\.\d+)?", i)
    #
    if '' in i:
        cent = re.findall(r"\d+(?:\.\d+)?\", i)
        total_cent = 0
        for k in cent:
            total_cent += float(re.findall('[0-9]+', k)[0])/100 
        dd.append('$'+str(total_cent))
    total_dol = 0
    for k in dd:
        total_dol += float(re.findall(r"[-+]?\d*\.\d+|\d+", k)[0])
     
    if total_dol != 0:
        print (i,':','$'+str(total_dol))
    else:
        print (i, ':')
    print('-'*50)
```

```python
def faceVal(X):
    c = X.split(" ")
    
    total_dol = 0
    for i in c:
        
        dd = re.findall(r"\$\d+(?:\.\d+)?", i)
       
        if '' in i:
            cent = re.findall(r"\d+(?:\.\d+)?\", i)
            total_cent = 0
            for k in cent:
                total_cent += float(re.findall('[0-9]+', k)[0])/100 
            dd.append('$'+str(total_cent))

        for k in dd:
            total_dol += float(re.findall(r"[-+]?\d*\.\d+|\d+", k)[0])

    if total_dol != 0:
        return '$'+str(total_dol)
    else:
        return 
```

```python
data['FaceVal'] = data['OfferDetails'].apply(lambda x: faceVal(x))
print (data.head(20))
```
